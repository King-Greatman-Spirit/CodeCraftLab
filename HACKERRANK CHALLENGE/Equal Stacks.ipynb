```python
function equalStacks(h1, h2, h3) {
    // Reverse elements in the stack
    h1 = h1.reverse();
    h2 = h2.reverse();
    h3 = h3.reverse();

    // Find sum of each stack
    let sum1 = h1.reduce((acc, height) => acc + height, 0);
    let sum2 = h2.reduce((acc, height) => acc + height, 0);
    let sum3 = h3.reduce((acc, height) => acc + height, 0);

    // Main logic
    while (true) {
        let minheight = Math.min(sum1, sum2, sum3);

        // Base case
        if (minheight === 0) {
            return 0;
        }

        // General case
        if (minheight < sum1) {
            sum1 -= h1.pop();
        }
        if (minheight < sum2) {
            sum2 -= h2.pop();
        }
        if (minheight < sum3) {
            sum3 -= h3.pop();
        }

        if (sum1 === sum2 && sum2 === sum3) {
            return sum1;
        }
    }
}

// Example usage
const h1 = [1, 2, 1, 1];
const h2 = [1, 2, 2];
const h3 = [1, 1];

const result = equalStacks(h1, h2, h3);
console.log(result);  // Output: 2
```
